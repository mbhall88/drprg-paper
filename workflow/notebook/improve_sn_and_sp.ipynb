```python
def split_var_name(name: str) -> tuple[str, int, str]:
    items = re.match(r"([A-Z]+)([-0-9]+)([A-Z/\*]+)", name, re.I).groups()
    return items[0], int(items[1]), items[2]

def is_indel(s):
    ref, pos, alt = split_var_name(s)
    return len(ref) != len(alt)

def is_tbp_indel(s):
    return "del" in s or "ins" in s or "dup" in s
```

```python
class MidpointNormalize(colors.Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))
```

```python
def comp(c1, c2):
    if pd.isna(c1):
        return c2
    elif pd.isna(c2):
        return c1
    elif c1 != c2:
        c1s = c1.split(";")
        c2s = c2.split(";")
        c3s = set(c1s).union(c2s)
        return ";".join(c3s)
    else:
        return c1
    
combine_func = lambda s1, s2: np.array([comp(c1, c2) for c1, c2 in zip(s1,s2)])
```

```python
from itertools import combinations
import re
from intervaltree import Interval
```

```python
summary = pd.read_csv("../../results/tables/sn_sp/summary.illumina.csv").set_index(["Drug", "Tool"], drop=False)
```

```python
rx = re.compile(r"(?P<lwr>\d+\.\d)-(?P<upr>\d+\.\d)%")
def extract_ci(s: str) -> tuple[float, float]:
    m = rx.search(s)
    if not m:
        raise ValueError(s)
    lwr = float(m.group("lwr"))
    upr = float(m.group("upr"))
    return lwr, upr
```

```python
drugs = set(summary["Drug"])
tools = set(summary["Tool"])
```

```python
results = []
for drug in drugs:
    for t1, t2 in combinations(tools, 2):
        t1_sn = summary.at[(drug, t1), "Sensitivity (95% CI)"]
        t1_sn_ci = extract_ci(t1_sn)
        t2_sn = summary.at[(drug, t2), "Sensitivity (95% CI)"]
        t2_sn_ci = extract_ci(t2_sn)
        
        sn_ci_overlaps = Interval(t1_sn_ci[0], t1_sn_ci[1] + 0.01).overlaps(t2_sn_ci[0], t2_sn_ci[1] + 0.01)
        if not sn_ci_overlaps:
            print(f"{drug} Sensitivity CI does not overlap for {t1} ({t1_sn_ci}) & {t2} ({t2_sn_ci})")
            results.append(["sensitivity", drug, t1, t1_sn_ci, t2, t2_sn_ci])
        
        t1_sp = summary.at[(drug, t1), "Specificity (95% CI)"]
        t1_sp_ci = extract_ci(t1_sp)
        t2_sp = summary.at[(drug, t2), "Specificity (95% CI)"]
        t2_sp_ci = extract_ci(t2_sp)
        
        sp_ci_overlaps = Interval(t1_sp_ci[0], t1_sp_ci[1] + 0.01).overlaps(t2_sp_ci[0], t2_sp_ci[1] + 0.01)
        if not sp_ci_overlaps:
            print(f"{drug} Specificity CI does not overlap for {t1} ({t1_sp_ci}) & {t2} ({t2_sp_ci})")
            results.append(["specificity", drug, t1, t1_sp_ci, t2, t2_sp_ci])
```

```python
print(pd.DataFrame(results, columns=["Metric", "Drug", "Tool1", "Tool1 CI", "Tool2", "Tool2 CI"]).to_markdown(index=False))
```
