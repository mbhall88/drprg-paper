```python
df = pd.read_csv("../../results/tables/sn_sp/classifications.nanopore.csv")
df.set_index(["run", "drug", "tool"], inplace=True, drop=False, verify_integrity=True)
df
```

```python
rf = 16
af = 10
rr = 16
ar = 24
r = rf+rr
a = af+ar
print(f"FRS {a/(r+a)}")
print(f"SB {min(af, ar)/(af+ar)}")
```

```python
stm_fp = set()
inh_fp = set()
eto_fp = set()
```

```python
for _, row in df.query("tool=='drprg' and classification=='FP'").iterrows():
    run = row["run"]
    drug = row["drug"]
    
    if drug != "isoniazid":
        continue
    
    mykrobe_clf = df.at[(run, drug, "mykrobe"), "classification"]
    tbp_clf = df.at[(run, drug, "tbprofiler"), "classification"]
    drprg_clf = df.at[(run, drug, "drprg"), "classification"]
    
    inh_fp.add(run)
    if not mykrobe_clf.startswith("F"):
        print(run, drug, "mykrobe")
    elif not tbp_clf.startswith("F"):
        print(run, drug, "tbprofiler")
    else:
        print(run, drug, "FN on all")
```
